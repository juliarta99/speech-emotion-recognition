```python
import os
import shutil
import glob
import math

# 1. Tentukan lokasi folder (UBAH BAGIAN INI)
source_dir = "dataset" 
base_dest_dir = "dataset_split"

# Ambil semua file .wav
audio_files = glob.glob(os.path.join(source_dir, "*.wav"))

# 2. Kumpulkan semua ID Aktor yang ada secara otomatis
aktor_unik = set()
for file_path in audio_files:
    filename = os.path.basename(file_path)
    parts = filename.split('_')
    if len(parts) >= 4:
        aktor_unik.add(int(parts[0]))

# Urutkan ID aktor dari yang terkecil ke terbesar
aktor_unik = sorted(list(aktor_unik))

# 3. Hitung proporsi 80% Train dan 20% Test
jumlah_aktor = len(aktor_unik)
jumlah_train = math.ceil(jumlah_aktor * 0.8)  # Dibulatkan ke atas untuk memastikan minimal 80%

# Pisahkan ID aktor ke dalam dua kelompok
aktor_train = set(aktor_unik[:jumlah_train])
aktor_test = set(aktor_unik[jumlah_train:])

print(f"Total Aktor Ditemukan: {jumlah_aktor} Orang")
print(f"Proporsi 80% Training: {len(aktor_train)} Aktor (ID {min(aktor_train)} s/d {max(aktor_train)})")
print(f"Proporsi 20% Testing : {len(aktor_test)} Aktor (ID {min(aktor_test)} s/d {max(aktor_test)})\n")
print("Memulai proses copy file ke folder train dan test...")

count_train = 0
count_test = 0

# 4. Mulai proses pemisahan dan pembuatan folder
for file_path in audio_files:
    filename = os.path.basename(file_path)
    parts = filename.split('_')
    
    if len(parts) >= 4:
        actor_id = int(parts[0])
        emotion = parts[2]
        
        # Cek apakah aktor ini masuk kelompok train atau test
        if actor_id in aktor_train:
            split_folder = "train"
            count_train += 1
        else:
            split_folder = "test"
            count_test += 1
            
        # Path folder tujuan: "hasil_split/train/SAD/"
        target_dir = os.path.join(base_dest_dir, split_folder, emotion)
        os.makedirs(target_dir, exist_ok=True)
        
        target_file_path = os.path.join(target_dir, filename)
        shutil.copy(file_path, target_file_path)

print("Proses Selesai!")
print(f"Total file di folder Training : {count_train} file")
print(f"Total file di folder Testing  : {count_test} file")
```
